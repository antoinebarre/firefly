# Error Models

```python
# IMPORT

import numpy as np
import matplotlib.pyplot as plt
```

```python
# GLOBAL PARAMETERS

t = np.arange(
    start=0,
    stop=100,
    step=1
)

real_values = -3. + 2.*t + 10*np.sin(2*np.pi/20*t)
```

```python
# TOOLS

def plot_comparison(
    t: np.ndarray,
    real_values: np.ndarray,
    measured_values: np.ndarray,
    title: str
    ) -> None :
    
    plt.figure()
    plt.title(title)
    plt.plot(t, real_values,'b', label='Real Values')
    plt.plot(t, measured_values,'r', label='Measured Values')
    plt.xlabel(r'$\tau$ [sec]')
    plt.ylabel('Values')
    plt.grid(True, which="both", ls="-", color='0.65')
    plt.legend()
    plt.show()




```

```python
# bias

measured_values = real_values + 10

plot_comparison(
    t=t,
    real_values=real_values,
    measured_values=measured_values,
    title= "Bias or Drift"
)
```

```python
# SCALE FACTOR

real_values = -200. + 4.*t

measured_values = real_values*(1 + 5E-1)
plot_comparison(
    t=t,
    real_values=real_values,
    measured_values=measured_values,
    title= "Scale Factor"
)
```
